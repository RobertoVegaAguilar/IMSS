```python
from datetime import date, timedelta

# Función para verificar si un año es bisiesto
def es_bisiesto(año):
    return (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0)

# Lista para almacenar los últimos días de cada mes
ultimos_dias_mes = []

# Rango de años desde 1998 hasta 2023
año_inicio = 1998
año_fin = 2023

for año in range(año_inicio, año_fin + 1):
    for mes in range(1, 13):  # 12 meses en un año
        # Determina el último día del mes teniendo en cuenta si es bisiesto o no
        if mes == 2 and es_bisiesto(año):
            ultimo_dia = 29
        elif mes == 2:
            ultimo_dia = 28
        elif mes in {4, 6, 9, 11}:  # Meses con 30 días
            ultimo_dia = 30
        else:  # Meses con 31 días
            ultimo_dia = 31
        
        # Agrega el último día del mes a la lista en formato 'YYYY-MM-DD'
        ultimos_dias_mes.append(date(año, mes, ultimo_dia).strftime('%Y-%m-%d'))

# Imprime la lista de los últimos días de cada mes
for fecha in ultimos_dias_mes:
    print(fecha)
```

```python
url_base = 'http://datos.imss.gob.mx/sites/default/files/asg-2023-01-31.csv'

# Crea una lista de 312 valores, todos iguales al URL base
lista_de_urls = [url_base] * 312

# Imprime la lista de URLs
for url in lista_de_urls:
    print(url)
```

```python
# Lista original
lista_original = lista_de_urls

# Lista de sustitución
lista_sustitucion = ultimos_dias_mes

# Sustituir la sección en cada componente de la lista original con las componentes de la lista de sustitución
for i in range(len(lista_original)):
    seccion_a_sustituir = lista_original[i][49:59]  # Define la sección a sustituir, por ejemplo, desde el índice 2 al 4
    lista_original[i] = lista_original[i].replace(seccion_a_sustituir, lista_sustitucion[i])

# Imprimir la lista original modificada
print(lista_original)
```
